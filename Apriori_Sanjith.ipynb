```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
data = pd.read_csv('/content/drive/MyDrive/online-retail-dataset.csv')
data
```

```python
data=data.dropna()
data
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
transactions = data.groupby(['InvoiceNo'])['Description'].apply(list).values.tolist()
results = list(apriori(transactions, min_support=0.01, min_confidence=0.1, min_lift=1, min_length=2))
```

```python
print("Frequent itemsets:")
for result in results:
    print("Items:", result.items)
    print("Support:", result.support)
    print("Confidence:", result.ordered_statistics[0].confidence)
    print()
```

```python
country_counts = data['Country'].value_counts()
country_with_max_transactions = country_counts.idxmax()
print("Country with the greatest number of transactions:", country_with_max_transactions)
```

```python
invoice_counts = data['InvoiceNo'].value_counts()
invoice_with_max_products = invoice_counts.idxmax()
print("Invoice number with the greatest number of products:", invoice_with_max_products)
```

```python
product_support = data['Description'].value_counts(normalize=True)
print("Support for each product:")
print(product_support)
```

```python
product_support = data['Description'].value_counts(normalize=True)
threshold = 0.002
frequent_products = product_support[product_support >= threshold]
print("Frequent products with support greater than or equal to", threshold)
print(frequent_products)
```

```python
basket=(data.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_sets=basket.applymap(lambda x: 1 if x > 0 else 0)
frequent_itemsets = fpgrowth(basket_sets, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(rules)
```
